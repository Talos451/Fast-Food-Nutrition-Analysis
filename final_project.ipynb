### Calories vs Sodium
This Python script visualizes the relationship between calories and sodium in menu items using a scatter plot.

Scatter Plot: Plots calories (x-axis) vs. sodium (y-axis).

```python
file_path = './NutritionMenu_Cleaned.csv' #Relative path. 
nutrition_data = pd.read_csv(file_path)

# Ensure relevant columns are properly loaded
calories = nutrition_data['Calories']
sodium = nutrition_data['Sodium (mg)']

# Set plot style and parameters
plt.rcParams['figure.facecolor'] = '#f7f7e8' 
plt.rcParams['axes.facecolor'] = '#f7f7e8'

# Create the regplot
plt.figure(figsize=(8, 6))
sns.regplot(x=calories, y=sodium, data=nutrition_data, 
            scatter_kws={'alpha': 0.9, 'color': '#c04040', 'edgecolors': 'w', 's': 80},
            line_kws={'color': '#5c6bc0', 'linewidth': 2})

# Customize the plot
plt.tick_params(colors='#5c6bc0')
plt.title('Relationship Between Calories and Sodium in Menu Items', fontsize=14)
plt.xlabel('Calories')
plt.ylabel('Sodium (mg)')

# Display the plot
plt.show()
```

### Is there a link between calories and sodium in the menu items?
There is a positive trend between calories and sodium, as higher-calorie items tend to have higher sodium content. However, numerous data points exist below the main trendline, suggesting outliers with lower sodium content relative to their calorie count. These are most likely drinks or other low-sodium items, which deviate from the general trend observed in solid food items.

### Nutritional Profiles Between Similar Items. 
This code filters similar items (e.g., burgers, fries, nuggets) from a menu dataset, groups them by restaurant chain, and calculates average nutritional values (calories, sodium, fat). It visualizes the results with two bar charts comparing calories and sodium across chains, styled with pastel colors and a beige background.

```python
keywords = [
    'burger', 'cheeseburger', 'double burger', 'big mac', 'whopper',
    'hamburger', 'fries', 'curly fries', 'wedges', 'potato', 'nuggets'
]
pattern = '|'.join(keywords)

# Filter the dataset for rows containing any of the keywords
filtered_data = nutrition_data[nutrition_data['Item'].str.contains(pattern, case=False, na=False)]

# Group by company and calculate mean nutritional values
nutrition_summary = filtered_data.groupby('Company')[['Calories', 'Sodium (mg)', 'Total Fat\n(g)']].mean()

# Plot the comparison for calories
plt.rcParams['figure.facecolor'] = '#f7f7e8' 
plt.rcParams['axes.facecolor'] = '#f7f7e8'
nutrition_summary['Calories'].plot(kind='bar', figsize=(8, 6), color='#90caf9', edgecolor='black')
plt.title('Average Calories in Similar Items Across Chains')
plt.xlabel('Company')
plt.ylabel('Calories')
plt.xticks(rotation=45)
plt.show()

# Plot the comparison for sodium
plt.rcParams['figure.facecolor'] = '#f7f7e8' 
plt.rcParams['axes.facecolor'] = '#f7f7e8'
nutrition_summary['Sodium (mg)'].plot(kind='bar', figsize=(8, 6), color='#f48fb1', edgecolor='black')
plt.title('Average Sodium in Similar Items Across Chains')
plt.xlabel('Company')
plt.ylabel('Sodium (mg)')
plt.xticks(rotation=45)
plt.show()
```

### How do similar items, like burgers or fries, compare in nutrition across chains?
McDonald's and Wendy's have the lowest sodium and calorie counts for similar items. KFC stands out with significantly higher sodium content, despite Burger King having higher calorie counts than KFC. This highlights notable differences in nutritional profiles across chains, even for comparable items.

### Keto Friendly Options
The following script defines keto-friendly criteria and selects all menu items that fit that criteria. 
Outputs a table that shows all keto friendly options. 

```python

# Define thresholds for keto-friendly criteria
high_fat_threshold = 20  # Example: 20g or more
low_carb_threshold = 50  # Adjusted to 50g

# Filter the dataset based on updated criteria
keto_friendly_items_updated = nutrition_data[
    (nutrition_data['Total Fat\n(g)'] > high_fat_threshold) &  # High fat
    (nutrition_data['Carbs\n(g)'] < low_carb_threshold)        # Low carbs
]

# Sort the results by Total Fat for better visualization
keto_friendly_items_updated_sorted = keto_friendly_items_updated.sort_values(by='Total Fat\n(g)', ascending=False)
keto_chart = keto_friendly_items_updated_sorted[['Item', 'Company', 'Calories', 'Total Fat\n(g)', 'Protein\n(g)', 'Carbs\n(g)']]
col_labels = ['Item', 'Company', 'Calories', 'Total Fat (g)', 'Protein (g)', 'Carbs (g)']

# Sort by calories for clear comparison
keto_chart_sorted = keto_chart.sort_values(by='Calories', ascending=False)

# Create a figure for the table
fig, ax = plt.subplots(figsize=(12, 6))
ax.axis('off') 

# Generate a table using matplotlib
table = ax.table(
    cellText=keto_chart_sorted.values,
    colLabels=col_labels,
    cellLoc='center',  # Center-align text
    loc='center',  # Place the table in the center of the axes
    colColours=["#f7d6d0"] * len(keto_chart_sorted.columns),  # Header colors
)

num_columns = len(keto_chart_sorted.columns)
column_indices = list(range(num_columns))
table.auto_set_column_width(column_indices)


# Display the table
plt.show()
```

### What are the keto-friendly options for all menu items? 
The table highlights keto-friendly menu options across different companies, sorted by calorie content. These items meet the criteria for high fat (>20g) and low carbohydrates (<50g), making them suitable for individuals following a ketogenic diet. The data provides a clear comparison of nutritional values, including total fat, protein, and carbs, to aid in informed decision-making. The criteria can be changed based on user requirements. 

All chains provide keto-friendly options based on our current criteria. There are high and lower calorie options to chose from. 

### Distribution of Saturated Fat Content Across Companies
This code extracts and groups saturated fat content by company, then visualizes the distribution using a boxplot. Allowing for comparison of saturated fat levels across chains. 

```python
# Group data by company
saturated_fat_by_company = nutrition_data[['Company', 'Saturated Fat\n(g)']]

# Display the grouped data
saturated_fat_by_company.head()

plt.figure(figsize=(12, 6))
sns.boxplot(
    data=saturated_fat_by_company,
    x='Company',
    y='Saturated Fat\n(g)',
    hue = 'Company',
    palette='pastel'
)
plt.legend = False
plt.title('Distribution of Saturated Fat Content Across Companies', fontsize=14)
plt.xlabel('Company')
plt.ylabel('Saturated Fat (g)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### How does saturated fat content vary between companies? 
McDonald's, Burger King, and Wendy's have similar average saturated fat levels, while Taco Bell's average is slightly lower. Pizza Hut stands out with the highest average saturated fat content, whereas KFC has the lowest. Notably, there are significant outliers for all companies except Pizza Hut, indicating a wider variation in saturated fat levels for most chains.

### Top 50 Items with the Highest Sodium Content.
This code identifies the top 50 menu items with the highest sodium content and visualizes them in a horizontal bar chart. A red dashed line marks the 2300 mg daily sodium limit for reference, making the chart particularly useful for individuals monitoring their sodium intake.

```python
top_sodium_items = nutrition_data.nlargest(50, 'Sodium (mg)')[['Item', 'Company', 'Sodium (mg)']]

# Create a horizontal bar chart
plt.figure(figsize=(12, 8))
ax = sns.barplot(
    data=top_sodium_items,
    y='Item',
    x='Sodium (mg)',
    hue='Company',
    palette='pastel',
    dodge=False
)

plt.axvline(x=2300, color='red', linestyle='--', linewidth=2, label='2300 mg Daily Maximum')

# Customize the plot
ax.set_title('Top 50 Items with the Highest Sodium Content', fontsize=14)
ax.set_xlabel('Sodium (mg)')
ax.set_ylabel('Item')
ax.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.tight_layout()
plt.show()
```

### What are the highest sodium menu items?
This table highlights the top 50 menu items with the highest sodium content. For individuals aiming to reduce their sodium intake, most of these items should be avoided—especially the five that exceed the daily maximum recommended intake. A more comprehensive list can be generated based on specific user needs.

### Which companies are overrepresented in the top 10 of the high-sodium menu items list?
KFC dominates the top 10 list, with 5 out of 10 items having the highest sodium content. Four of these items exceed the daily maximum recommended intake. Notably, at least three of these items include "family" in their name, suggesting they are intended to be shared rather than consumed by a single person. On the other hand, Taco Bell is underrepresented on this list, only appearing in the 49th and 50th spots.
